## Matplotlib library examples 

** Part 1 Basics ** 

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np 
x = np.linspace(0,5,11)
y = x**2
```

```python
# Functional method 
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Parabolic')
```

```python
plt.subplot(1,2,1)   # subplot(length, breath, plot number)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object oriented method 
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,.8])  # % of start x coordinate y coordinate, width and height 
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])  # axes are [left,bottom, width, height]
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
```

** Part 2  working with subplots** 

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
for current_ax in axes: 
    current_ax.plot(x,y)

plt.tight_layout()
#axes[0].plot(x,y) also does the sames thing.
#axes[1].plot(x,y)
```

** Figure Size and DPI ** 

```python
fig = plt.figure(figsize=(3,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(4,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x square')
ax.plot(x,x**3,label='x cubed')   # adding to plots to the same figure leads to two graphs on the figure as can be seen. 
ax.legend(loc=0) # best position for the legend
```

```python
# saving a figure 
fig.savefig('my-fig.png', dpi=100)
```

** Customize the look and feel ** 

```python
fig = plt.figure(figsize=(3,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth=3,linestyle='-', 
        marker='o', markersize=10, markerfacecolor='yellow', markeredgewidth='3')  # we can get rgb hex codes as well. 
```

```python
fig = plt.figure(figsize=(3,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth=3,linestyle='--')
```
