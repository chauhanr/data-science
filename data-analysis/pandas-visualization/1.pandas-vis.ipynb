## Pandas Visualization 

```python
import numpy as np 
import pandas as pd
import seaborn as sns
%matplotlib inline 
```

```python
df1 = pd.read_csv('df1', index_col=0)
df1.head()
```

```python
df2 = pd.read_csv('df2')
df2.head()
```

```python
df1['A'].hist(bins=30)
```

```python
df1['A'].plot(kind='hist', bins=30)
```

```python
df2.plot.area(alpha=0.4)  # area plot with transparency.
```

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

```python
df1.plot.line(x=df1.index,y='B', lw=1)
```

```python
df1.plot.scatter('A','B', c='C',cmap='coolwarm')
```

```python
df2.plot.box()
```

```python
df = pd.DataFrame(np.random.randn(1000,2), columns=['a','b'])
df.head()
```

```python
df.plot.hexbin(x='a',y='b',gridsize=25)
```

```python
df2['a'].plot(kind='kde')  # kernel density estimation.
```

```python
df2.plot.density()  # kde for all columns are given here. 
```
