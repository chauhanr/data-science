## Matric Plot 

```python
import seaborn as sns 
%matplotlib inline 
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr()
tc
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
fc = flights.pivot_table(index='month', columns='year', values='passengers')
fc
```

```python
sns.heatmap(fc, cmap='coolwarm', linecolor='black',linewidths=1)
```

```python
sns.heatmap(fc, cmap='coolwarm')
```

```python
sns.clustermap(fc, cmap='coolwarm', standard_scale=1)
```

## Regression Plot 
** Examples of regression plot ** 

```python
# lm model will give linear model 
sns.lmplot(x='total_bill',y='tip',data=tips, hue='sex', markers=['o','v'],scatter_kws={'s':50})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex', aspect=0.6, size=8)
```
