## Seaborn Library 
** Distribution Plots ** 

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'])  # historgram that gives us a distribution 
```

```python
sns.distplot(tips['tip'])
```

```python
sns.jointplot(x= 'total_bill',y= 'tip', data=tips, kind='hex') # hexagonal representation 
```

```python
sns.pairplot(data=tips, hue='time')  
```

```python
sns.rugplot(tips['total_bill'])  # different representation of distribution. 
```

```python
sns.kdeplot(tips['total_bill'])
```
