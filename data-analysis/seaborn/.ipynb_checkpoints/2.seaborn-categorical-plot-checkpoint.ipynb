## Categorical Data Plots 

```python
import seaborn as sns
import numpy as np
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex', y='total_bill',data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)  # just counts the number of the x axis column choosen therefore there is no y.
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', split=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
# drawback - when data gets more they are not very good at visualization. 
# swarm plot are not very popular. 
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')

```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
# generic plot method and we send a kind of the plot.
```
